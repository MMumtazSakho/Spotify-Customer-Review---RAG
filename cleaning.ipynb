```python
import gc
import os
import pickle
import gc  

def save_embeddings(embeddings, step):
    if not os.path.exists('embeddings_openai2'):
        os.makedirs('embeddings_openai2')
    
    file_name = f'embeddings_openai2/embedding_{step}.pkl'
    with open(file_name, 'wb') as f:
        pickle.dump(embeddings, f)

def save_texts(texts, step):
    if not os.path.exists('embeddings_openai2'):
        os.makedirs('embeddings_openai2')
    
    file_name = f'embeddings_openai2/texts_{step}.pkl'
    with open(file_name, 'wb') as f:
        pickle.dump(texts, f)

def batch_request(client, list_text, embedding_model_name, batch_size=1000):
    embeddings = []  
    texts = []
    total_requests = len(list_text) // batch_size + 1
    save_step =  100 
    batch_counter = 0  

    for i in range(total_requests):
        start_idx = i * batch_size
        end_idx = min((i + 1) * batch_size, len(list_text))

        batch = list_text[start_idx:end_idx]

       
        embeddings_batch_response = client.embeddings.create(input=batch, model=embedding_model_name)

        embeddings_batch = [embeddings_batch_response.data[i].embedding for i in range(len(embeddings_batch_response.data))]

        embeddings.extend(embeddings_batch)
        texts.extend(batch)
        print(f'Request ke-{i+1} dari {total_requests}')

        if (i + 1) % save_step == 0 or (i + 1) == total_requests:
            batch_counter += 1
            save_embeddings(embeddings, batch_counter)
            save_texts(texts, batch_counter)
            print(f'Embeddings disimpan pada file embedding_{batch_counter}.pkl')
            print(f'Texts disimpan pada file texts_{batch_counter}.pkl')

            del embeddings
            del texts
            gc.collect()  
            embeddings = []
            texts = []

    return embeddings
```

```python
embeddings = batch_request(client, list_text, embedding_model_name, batch_size=1000)
```
